## Redução de Dimensionalidade em Imagens para Redes Neurais

Projeto do curso BairesDev - Machine Learning Practitioner na plataforma DIO

```python
# Import das bibliotecas

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Funções para transformar uma imagem colorida para níveis de cinza (0 a 255) e para binarizada (0 e 255), preto e branco

def transfCinva(imagem):
  alt, larg, _ = imagem.shape
  img_cinza = np.zeros((alt, larg), dtype=np.uint8)

  for i in range(alt):
    for j in range(larg):
      r, g, b = imagem[i, j]
      img_cinza[i, j] = int(0.299 * r + 0.587 * g + 0.114 * b)

  return img_cinza

def transfBinarizada(img_cinza, limiar):
  alt, larg = img_cinza.shape
  img_bin = np.zeros((alt, larg), dtype=np.uint8)

  for i in range(alt):
    for j in range(larg):
      if img_cinza[i, j] >= limiar:
        img_bin[i, j] = 255
      else:
        img_bin[i, j] = 0

  return img_bin


def plotImgs(imagem, img_cinza, img_bin):
  plt.figure(figsize=(15, 5))
  plt.subplot(1, 3, 1)
  plt.title("Imagem Original")
  plt.imshow(imagem)
  plt.axis("off")

  plt.subplot(1, 3, 2)
  plt.title("Imagem Cinza")
  plt.imshow(img_cinza, cmap="gray")
  plt.axis("off")

  plt.subplot(1, 3, 3)
  plt.title("Imagem Binarizada")
  plt.imshow(img_bin, cmap="gray")
  plt.axis("off")

  plt.show()


caminhoImagem = '/content/imagem_teste.png'
limiar = 128

imagem = cv2.imread(caminhoImagem)
img_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
img_cinza = transfCinva(img_rgb)
img_bin = transfBinarizada(img_cinza, limiar)

plotImgs(img_rgb, img_cinza, img_bin)
```
